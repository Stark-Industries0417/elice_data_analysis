# 상점 이름 가져오기

```python
import pandas as pd
```

```python
kakao = pd.read_csv("../kakao/kakao_restaurant.csv", header=None, names=["name", "s", "t", "c", "u"])
kakao.head()
```

```python
kakao_name = list(kakao["name"])
kakao_name
```

```python
naver = pd.read_csv("../naver/네이버 업소명 총 평점 업종 url.csv")
naver.head()
```

```python
naver_name = list(naver["업소명"])
naver_name
```

```python
intersection_restaurant = [f"선릉역 음식점 {name}" for name in list(set(kakao_name).intersection(set(naver_name)))]
intersection_restaurant
```

```python
with open("intersection_restaurant.py", "w") as output:
    output.write("intersection_restaurant = ")
    output.write(str(intersection_restaurant))
```
